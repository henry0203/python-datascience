```python
group_age=black_friday.groupby("Age")
group_age
```

```python
group_age.count()
```

```python
customer_buy=dict()
for age, group in group_age:
    customer_buy[age]=sum(group["Purchase"])
    #print(age) #category name
    #print(group) #subdataframe
```

```python
customer_buy
```

```python
plt.plot(list(customer_buy.keys()), list(customer_buy.values()))
plt.draw()
#list(customer_buy.keys())
```

```python
plt.plot(list(customer_buy.keys()), list(customer_buy.values()), marker='x')
plt.draw()
```

```python
plt.plot(list(customer_buy.keys()), list(customer_buy.values()), marker='o',linestyle="")
plt.xlabel("age categories")
plt.ylabel("Sales sum")
plt.title("Total sales/ age categories", y=1.05)
plt.draw()
```

```python
plt.rcParams["figure.figsize"]=[16,9]
plt.scatter(list(customer_buy.keys()), list(customer_buy.values()), c=list(customer_buy.values()), cmap="seismic", marker='D', s=100) # matplolib color
plt.xlabel("age categories")
plt.ylabel("Sales sum")
plt.title("Total sales/ age categories", y=1.05)
plt.colorbar()  #vertical by default, set in param orientation="horizontal" if horizontal
plt.draw()
plt.savefig("mon_graphique.png", bbox_inches="tight")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas
tips=sns.load_dataset('tips')
tips
```

```python
# 1. Tracer un barplot avec la fonction sns.barplot reprÃ©sentant l'addition totale payÃ©e et en sÃ©parant ces donnÃ©es par sexe
sns.barplot(x="sex", y="total_bill", data=tips)
plt.draw()
```

```python
# 2. Etudier la distribution de la variable total_bill grÃ¢ce Ã  la fonction distplot
sns.distplot(tips["total_bill"], kde=False) #input serie => one column
plt.draw()
```

```python
# 3.Utiliser la fonction jointplot de seaborn pour tracer la variable total_bill (X) en fonction de la variable tip (Y)
sns.jointplot(x="total_bill", y="tip", data=tips, kind="reg")
plt.draw()
```

```python
# 4. Utiliser la fonction pairplot de Seaborn pour Ã©tudier les relations entre nos variables
sns.pairplot(tips)
plt.draw()
```

```python
# 5. Tracer le boxplot de la variable total_bill en crÃ©ant un boxplot par jour (day) 
# et en utilisant la variable sex dans l'option hue
sns.boxplot(x="day", y="total_bill", hue="sex", data=tips)
plt.draw()
```
